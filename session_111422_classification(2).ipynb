# 1. Lightgbm

* 데이터 '와인품질분류.zip'입니다.

1. 필요한 라이브러리를 가져옵니다.

```python
import pandas as pd
import lightgbm as lgbm
```

2. 데이터를 불러옵니다.

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
smpl_sub = pd.read_csv('data/sample_submission.csv')
```

```python
train.head()
```

```python
test.head()
```

3. 변수를 변환합니다. 

```python
# type에는 white와 red 두 종류가 있습니다.
# 각각 0,1로 변환합니다.
train['type'].replace("white", 0, inplace=True) 
train['type'].replace("red", 1, inplace=True) 
test['type'].replace("white", 0, inplace=True) 
test['type'].replace("red", 1, inplace=True) 

train['type'].replace({'white':0, 'red':1}, inplace=True)
test['type'].replace({'white':0, 'red':1}, inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
# 모델에 입력하기 전 데이터를 정형합니다.
# train_x : train의 'index', 'quality' 피처를 제거하고 저장합니다.
# train_y : train의 'quality'를 저장합니다.
# test_x : test의 'index'를 제거하고 저장합니다.
train_x = train.drop(['quality'], axis=1)
train_y = train['quality']
test_x = test
```

```python
train_x.shape, train_y.shape, test_x.shape
```

```python
train_x.head() 
```

```python
train_y.head()
```

```python
head(train_y)
```

```python
test_x.head()
```

4. LightGBM 모델을 생성 및 훈련합니다.

```python
# LightGBM의 파이썬 패키지인 lightgbm에서 LGBMClassifier Import
from lightgbm import LGBMClassifier

lgbm_wrapper = LGBMClassifier()
lgbm_wrapper.fit(train_x,train_y)
```

5. 생성한 모델로 예측 데이터를 생성합니다.

```python
y_pred = lgbm_wrapper.predict(test_x)
```

6. 제출파일을 생성합니다.

```python
smpl_sub.head() # 왜 head(smpl_sub)는 오류가 나는지...
```

```python
#서브미션 파일의 quality 변수에 예측값을 대입합니다.
smpl_sub['quality'] = y_pred
```

```python
smpl_sub.head()
```

```python
# csv 파일로 저장합니다.
smpl_sub.to_csv('submission.csv', index=False)
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.head()
```

```python
test.head()
```

```python
# 원핫 인코딩 (pd.get_dummies())
train_one = pd.get_dummies(train)
test_one = pd.get_dummies(test)
```

```python
# 모델 정의
import xgboost as xgb
from xgboost import XGBClassifier

model = XGBClassifier()
```

```python
# 모델 학습

# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수

X = train_one.drop('quality', axis=1)
y = train_one['quality']
# fit 메소드를 이용해 모델 학습
model.fit(X,y)
```

```python
# predict 메소드와 test_one 데이터를 이용해 품질 예측
y_pred = model.predict(test_one)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기
smpl_sub = pd.read_csv('data/sample_submission.csv')
```
